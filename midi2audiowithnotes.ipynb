```python
# --- SETUP ---
from music21 import stream, note, tempo, midi
from midi2audio import FluidSynth
import IPython.display as ipd
```

```python
# --- CREATE MELODY ---
melody = stream.Stream()
melody.append(tempo.MetronomeMark(number=120))
```

```python
# Add some notes (C major scale)
notes = ['C4', 'D4', 'E4', 'F4', 'G4', 'A4', 'B4', 'C5']
for pitch in notes:
    melody.append(note.Note(pitch, quarterLength=0.5))
```

```python
# --- SAVE AS MIDI ---
midi_fp = "/content/melody.mid"
mf = midi.translate.music21ObjectToMidiFile(melody)
mf.open(midi_fp, 'wb')
mf.write()
mf.close()

```

```python
# --- PLAY MIDI IN COLAB ---
ipd.Audio(midi_fp)

```

```python
# Download a soundfont file
!wget -O /content/sf2.sf2 https://musical-artifacts.com/artifacts/1990/download -q

# Convert MIDI to WAV
fs = FluidSynth(sound_font='/content/sf2.sf2')
fs.midi_to_audio(midi_fp, '/content/melody.wav')

# Play WAV
ipd.Audio('/content/melody.wav')
```
